```python
def fact(n):
  faktorijel = 1
  for i in range(n):
    faktorijel *= i
  return faktorijel

def choose(n,k):
  rez = fact(n) / (fact(k)*fact(n-k))

  return rez

def main():
  n = int(input("Unesi n: "))
  k = int(input("Unesi k: "))
  rezz = choose(n,k)
  print(rezz)
```

```python
import numpy as np

def kvadratna(M):
    duzina = len(M)

    n = int(np.sqrt(duzina))

    if (n * n == duzina):
        return M.reshape((n, n))
    else:
        return None
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 101)

x = np.sin(n)
y = np.cos(n)

plt.subplot(2, 1, 1)
plt.plot(n, x)
plt.title('x = sin(n)')
plt.xlabel('n')
plt.ylabel('sin(n)')
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(n, y)
plt.title('y = cos(n)')
plt.xlabel('n')
plt.ylabel('cos(n)')
plt.grid(True)

plt.show()
```
